# 内容推荐概念

   关联规则：事物之间的相互依存性或关联性，以此可以发现事物之间类似啤酒和尿布的内在关系
   
   项集：在一个交易中，包含的所有项目集合，如订单所包含的所有种类商品
   
   支持度：某个商品组合出现的概率，次数和总次数的比例
   
   频繁项集：某个项集的支持度大于设定阈值（人为设定或者根据数据分布和经验来设定），即称这个项集为频繁项集
   
   阈值的设置方法：1：经验值 2：根据支持度排序取前面指定数量的最后一个支持度
   
   置信度：一种条件概率，当用户购买了X有多大概率购买产品Y ｛X->Y ｝ 的概率不一定等于｛X->Y ｝的概率
   
   提升度：置信度{X->Y}/支持度X,只有比值大于1的时候表示X对Y才有提升效果，等于1表示没有提升效果，小于1表示，事物内部出现了一定的竞争关系

# 回归分析

     回归模型的评价指标：MAE MSE RMSE和R方值,而分类问题我们使用的是准确率评估(值在0-1之间)
    
     MAE(平均绝对误差): 所有预测值与真实值之间的距离的绝对值之和的平均值,越接近于0，说明模型选择和拟合更好，数据预测也越成功
     其公式为：
    
     

   MSE(均方误差) 用 真实值-预测值 然后平方之后求和平均,与线性函数的损失值类似，

    以上两个算法都有写局限性，同一个算法模型，解决不同的问题，不能体现此模型针对不同问题所表现的优劣。因为不同实际应用中，数据的量纲不同，无法直接比较预测值，因此无法判断模型更适合预测哪个问题
    
    R方值 将预测结果转换为准确度，来表示模型对现实数据拟合的程度，评估预测效果.它是用1-预测模型/基准模型
    

对公式的理解：

    分子是我们的模型预测产生的错误，分母是使用y等于y的均值这个模型所产生的错误
    
    自己的模型预测产生的错误 / 基础模型预测生产的错误，表示自己的模型没有拟合住的数据，因此$R^2$可以理解为，自己的模型拟合住的数据
    
    基准模型问题：因为其没有考虑x的取值，只是很生硬的以为所有的预测样本，其预测结果都是样本均值

R方模型结论:

   $R^2$ <= 1
      
   $R^2$越大越好，当自己的预测模型不犯任何错误时：$R^2$ = 1

   当我们的模型等于基准模型时：R2 = 0

   如果R2 < 0，说明学习到的模型还不如基准模型。  # 注：很可能数据不存在任何线性关系
